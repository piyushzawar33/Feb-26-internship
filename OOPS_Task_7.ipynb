```python
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.is_on = False  # Initial status is OFF

    def turn_on(self):
        self.is_on = True

    def turn_off(self):
        self.is_on = False

    def get_status(self):
        return f"{self.name} is {'ON' if self.is_on else 'OFF'}"

# Sample Input
light_name = "Bedroom Light"
action = "ON"

# Create a smart light instance
bedroom_light = SmartLight(light_name)

# Perform action
if action == "ON":
    bedroom_light.turn_on()
elif action == "OFF":
    bedroom_light.turn_off()

# Expected Output
print(bedroom_light.get_status())
```

```python
class Employee:
    def __init__(self, name, employee_id, department):
        self.name = name
        self.employee_id = employee_id
        self.department = department

    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.employee_id}")
        print(f"Department: {self.department}")

# Sample Input
employee_name = "Rahul"
employee_id = "EMP102"
employee_department = "IT"

# Create an employee instance
rahul = Employee(employee_name, employee_id, employee_department)

# Display ID card details
rahul.display_id_card()
```

```python
class Contact:
  def __init__(self, name, phone_no):
    self.name = name
    self.phone_no = phone_no

  def display_information(self):
    print("Contact Saved")
    print(f"Name: {self.name}")
    print(f"Phone Number: {self.phone_no}")

contact_name = "Anita"
phone_no = "9876543210"

Anita = Contact(contact_name, phone_no)
Anita.display_information()
```

```python
class PriceTag:
  def __init__(self, name, price):
    self.name = name
    self.price = price

  def display_tag(self):
    print(f"Product: {self.name}")
    print(f"Price:  ₹{self.price}")

product_name = "Headphones"
price = 2499


PriceTag(product_name, price).display_tag()
```

```python
class Rating:
  def __init__(self, name, rating):
    self.name = name
    self.rating = rating

  def display_details(self):
    print(f"Movie: {self.name}")
    print(f"Rating: {self.rating}/5")

movie_name = "Inception"
rating = 4.8

Rating(movie_name, rating).display_details()
```

```python
class Delivery:
  def __init__(self, name, address):
    self.name = name
    self.address = address

  def display_details(self):
    print("Delivery Details")
    print(f"Customer: {self.name}")
    print(f"Address: {self.address}")

customer_name = "Suman"
delivery_address = "Hyderabad"

Delivery(customer_name, delivery_address).display_details()
```
