```python
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Identify the highest performance score
highest_score = 0
if employees:
    highest_score = max(employees.values())

# Handle ties and display all employees eligible for the top performance bonus
top_performers = []
for employee, score in employees.items():
    if score == highest_score:
        top_performers.append(employee)

# Format the output
if top_performers:
    performers_str = ", ".join(top_performers)
    print(f"Top Performers Eligible for Bonus: {performers_str} (Score: {highest_score})")
else:
    print("No employees found or no top performers identified.")
```

```python
import collections
import re

# Input search query
search_query = "Buy mobile phone buy phone online"

# Convert to lowercase
search_query_lower = search_query.lower()

# Ignore common punctuation and split into words
cleaned_query = re.sub(r'[.,;!\"\'?]', ' ', search_query_lower)
words = cleaned_query.split()

# Count the frequency of each keyword
keyword_counts = collections.Counter(words)

# Display only keywords searched more than once
result = {word: count for word, count in keyword_counts.items() if count > 1}

print(result)
```

```python
# Input
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid sensor readings
valid_readings = []

# Iterate through readings with their indices
for hour_index, reading_value in enumerate(sensor_readings):
    # Check if the reading is an even number
    if reading_value % 2 == 0:
        # Store as (hour_index, reading_value) pair
        valid_readings.append((hour_index, reading_value))

# Display the result
print(f"Valid Sensor Readings (Hour, Value):\n{valid_readings}")
```

```python
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Count how many users belong to each email domain
domain_counts = collections.defaultdict(int)
for email in emails:
    # Extract domain from email (e.g., 'gmail.com' from 'ravi@gmail.com')
    domain = email.split('@')[1]
    domain_counts[domain] += 1

# Calculate the total number of emails
total_emails = len(emails)

# Calculate and display the percentage usage of each domain
if total_emails > 0:
    print("Email Domain Usage Analysis:")
    for domain, count in domain_counts.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")
else:
    print("No emails provided for analysis.")
```

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate the average daily sales
if sales:
    average_sales = sum(sales) / len(sales)
else:
    average_sales = 0

# Detect days where sales are more than 30% above average
spike_threshold = average_sales * 1.30

print("Sales Spike Detection:")
found_spike = False
for day_index, sale_value in enumerate(sales):
    if sale_value > spike_threshold:
        print(f"Day {day_index + 1}: {sale_value}")
        found_spike = True

if not found_spike and sales:
    print("No sales spikes detected above 30% of the average.")
elif not sales:
    print("No sales data provided.")
```

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Count the frequency of each user ID
id_counts = collections.Counter(user_ids)

# Display only duplicate user IDs and their counts
print("Duplicate User ID Detection:")
found_duplicates = False
for user_id, count in id_counts.items():
    if count > 1:
        print(f"{user_id} → {count} times")
        found_duplicates = True

if not found_duplicates:
    print("No duplicate user IDs found.")
```
